```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
original_path = '/Users/dada/Desktop/TFM/EARLINET_Database/FINAL_CASES/BCN_12_04_2024/TO_BE_TESTED_VSD_BCN12thApril2024/Original.csv'
avg_data_path = '/Users/dada/Desktop/TFM/EARLINET_Database/FINAL_CASES/BCN_12_04_2024/TO_BE_TESTED_VSD_BCN12thApril2024/Realization_with_avg_values_from_realizations.csv'
```

```python
original_df = pd.read_csv(original_path)
avg_data_df = pd.read_csv(avg_data_path)

radii = original_df['r(um)']
original_vsd = original_df['VSD(um^3/cm^3)']
original_std = original_df['VSD_std(um^3/cm^3)']

avg_vsd = avg_data_df['VSD(um^3/cm^3)']
avg_std = avg_data_df['VSD_std(um^3/cm^3)']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(radii, original_vsd, color='r', label='Original VSD')
plt.plot(radii, avg_vsd, color='b', label='VSD calculated')
# plt.errorbar(radii, original_vsd, yerr=original_std, fmt='.', capsize=5, capthick=1, ecolor='red', label='Original')
# plt.errorbar(radii, avg_vsd, yerr=avg_std, fmt='.', capsize=1, capthick=1, ecolor='blue', label='Calculated')
plt.xscale('log')
plt.title('Plot of Volume Size Distribution')
plt.xlabel('Radius (micrometers)')
plt.ylabel('VSD')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(radii, original_std, color='r', label='Original errors')
plt.plot(radii, avg_std, color='b', label='errors calculated')
# plt.errorbar(radii, original_vsd, yerr=original_std, fmt='.', capsize=5, capthick=1, ecolor='red', label='Original')
# plt.errorbar(radii, avg_vsd, yerr=avg_std, fmt='.', capsize=1, capthick=1, ecolor='blue', label='Calculated')
plt.xscale('log')
plt.title('Plot of VSD Standadard Deviation')
plt.xlabel('Radius (micrometers)')
plt.ylabel('STD')
plt.grid(True)
plt.legend()
plt.show()
```
